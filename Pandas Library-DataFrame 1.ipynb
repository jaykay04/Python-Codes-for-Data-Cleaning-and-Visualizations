```python
#Drop Columns in Dataframe using column label or position
#Lets create a DataFrame
cars_data=[('benz',2018,'New York','USA',10500),
           ('bmw',2020,'Madrid','Spain',9500),
           ('Gwagon',2019,'Berlin','Germany',11000),
           ('Bentley',2021,'Washington','USA',12000),
           ('Ferrari',2017,'Tokyo','Japan',12500)
          ]
cars_df=pd.DataFrame(cars_data,columns=['cars','year','city','coutry','price'],index=['a','b','c','d','e'])
cars_df
```

```python
#Delete single column by column name
cars_mod=cars_df.drop('city',axis='columns')
cars_mod
```

```python
#Delete multiple columns by label name
cars_mod2=cars_df.drop(['coutry','price'],axis='columns')
cars_mod2
```

```python
#Drop columns by Index Position
cars_mod3=cars_df.drop([cars_df.columns[1],cars_df.columns[2]],axis='columns')
cars_mod3
```

```python
#Using the inplace function
cars_df.drop(['cars','price'],axis='columns',inplace=True)
cars_df
```

```python
#Drop Columns using if exist
if 'city' in cars_df.columns:
    cars_df.drop(['city'],axis='columns',inplace=True)
else:
    print('Column does not exist')
```
