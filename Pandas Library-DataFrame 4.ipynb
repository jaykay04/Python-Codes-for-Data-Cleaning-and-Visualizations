```python
import numpy as np
import pandas as pd
#Concatenating Dataframes along rows and columns
df1=pd.DataFrame({'A':['A0','A1','A2','A3'],
                  'B':['B0','B1','B2','B3'],
                  'C':['C0','C1','C2','C3'],
                  'D':['D0','D1','D2','D3']
                 })
df1
```

```python
df2=pd.DataFrame({'A':['A4','A5','A6','A7'],
                       'B':['B4','B5','B6','B7'],
                       'C':['C4','C5','C6','C7'],
                       'D':['D4','D5','D6','D7']},
                      index=[4,5,6,7])
df2
```

```python
df3=pd.DataFrame({'A':['A8','A9','A10','A11'],
                       'B':['B8','B9','B10','B11'],
                       'C':['C8','C9','C10','C11'],
                       'D':['D8','D9','D10','D11']},
                      index=[8,9,10,11])
df3
```

```python
#Concatenating along rows
pd.concat([df1,df2,df3])
```

```python
#Concatenating along columns
pd.concat([df1,df2,df3], axis=1)
```

```python
#Joining Dataframes
data_df1=pd.DataFrame({'A':['A0','A1','A2','A3'],
                       'B':['B0','B1','B2','B3']},
                      index=['K0','K1','K2','K3'])

data_df2=pd.DataFrame({'C':['C0','C1','C2','C3'],
                       'D':['D0','D1','D2','D3']},
                      index=['K0','K1','K2','K3'])
print(data_df1)
print(data_df2)
```

```python
data_df1.join(data_df2) #Join occurs on the basis of index only.
```

```python
#Merging Dataframes
data1=pd.DataFrame({'Key':['K0','K1','K2','K3'],
                       'A':['A0','A1','A2','A3'],
                       'B':['B0','B1','B2','B3']})

data2=pd.DataFrame({'Key':['K0','K1','K2','K3'],
                       'A':['A0','A1','A2','A3'],
                       'B':['B0','B1','B2','B3']})

print(data1)
print(data2)
```

```python
pd.merge(data1,data2,how='inner',on='Key')
```

```python
#Lets look at more merge operations
employees= [ (11, 'jack', 34, 'Sydney', 5) ,
           (12, 'Riti', 31, 'Delhi' , 7) ,
           (13, 'Aadi', 16, 'New York', 11) ,
           (14, 'Mohit', 32,'Delhi' , 15) ,
           (15, 'Veena', 33, 'Delhi' , 4) ,
           (16, 'Shaunak', 35, 'Mumbai', 5 ),
            (17, 'Shaun', 35, 'Colombo', 11)
            ]
emp_df = pd.DataFrame(employees,columns=['ID','Name','Age','City','Experience'],
                     index=['a','b','c','d','e','f','g'])
emp_df
```

```python
salaries = [(11, 5, 70000, 1000) ,
           (12, 7, 72200, 1100) ,
           (13, 11, 84999, 1000) ,
           (14, 15, 90000, 2000) ,
           (15, 4, 61000, 1500) ,
           (16, 5, 71000, 1000),
           (21, 10,81000, 2000)
            ]
sal_df=pd.DataFrame(salaries,columns=['ID','Experience','Salary','Bonus'],
                        index=['a','b','c','d','e','f','g'])
sal_df
```

```python
#Merge on  common columns using default inner join
merge_df1=emp_df.merge(sal_df)# This automatically does inner join and on common columns(keys)i.e'ID' & 'Experience'
merge_df1                     
```

```python
pd.merge(emp_df,sal_df, how='inner') #Inner join picks the common rows in both dataframes
```

```python
#Lets do a left join
merge_df2=emp_df.merge(sal_df,how='left') #This include rows on the left dataframe and add NAN for values
                                          #for the missing values in the right dataframe for those columns(keys)
merge_df2
```

```python
#Right Join
merge_df3=emp_df.merge(sal_df,how='right') #This include rows on the right dataframe and add NAN for values
                                          #for the missing values in the left dataframe for those columns(keys)
merge_df3
```

```python
#Outer Join
merge_df4=emp_df.merge(sal_df,how='outer') #This include all rows from right and left dataframes and add NAN
                                           #for values which are missing in either left or right dataframe
                                           #for any key.
merge_df4
```

```python
#Lets Merge on specific columns
merge_df5=emp_df.merge(sal_df,how='inner',on='ID')
merge_df5
```

```python
#Using suffixes for similar column names
merge_df6=emp_df.merge(sal_df,how='inner',on='ID',suffixes=('_In_Years','_Levels'))
merge_df6
```

```python
#Merging dataframes with different columns
#Lets change the ID column to empID in sal_df
sal_df.rename(columns={'ID':'EmpID'},inplace=True)
sal_df
```

```python
#Lets merge on key 'ID' from emp_df and key 'EmpID' on sal_df
merge_df7=emp_df.merge(sal_df,left_on='ID',right_on='EmpID')
merge_df7
```

```python
#OUTER MERGE
player_file1.merge(player_file2,on='Player',how='outer')
```

```python
#Count NaN or Missing values in Dataframes
students = [ ('jack', np.NaN, 'Sydeny' , 'Australia') ,
                 ('Riti', np.NaN, 'Delhi' , 'India' ) ,
                 ('Vikas', 31, np.NaN , 'India' ) ,
                 ('Neelu', 32, 'Bangalore' , 'India' ) ,
                 ('John', 16, 'New York' , 'US') ,
                 ('John' , 11, np.NaN, np.NaN ) ,
                (np.NaN , np.NaN, np.NaN, np.NaN ) 
                 ]
stud_df=pd.DataFrame(students, columns = ['Name' , 'Age', 'City' , 'Country'])
stud_df
```

```python
stud_df.isnull()#This returns true for where there are NaN values
```

```python
#Count all NaN values for both rows and columns
stud_df.isnull().sum().sum()
```

```python
#Count NaN values column-wise
stud_df.isnull().sum()
```

```python
#Count NaN values row-wise
#To do this, we have to iterate over each row and call and call sum() on it
for i in range(len(stud_df.index)):
    print('NaN in row',i,':',stud_df.iloc[i].isnull().sum())
```

```python
#Count number of rows using .shape and .index
#shape
stud_df.shape[0]
```

```python
#Mapping in DataFrame:This operation is used to add columns to dataframe and fill it with values
country_df=pd.DataFrame({'Country':['Afghanistan','Algeria','Albania'],
                         'Code':[93,355,213]
                        })
country_df
```

```python
GDP_dict={'Afghanistan':20,'Albania':12.8,'Algeria':215}
#Lets map
country_df['GDP']=country_df['Country'].map(GDP_dict)
country_df
```

```python
info={'First_name': ['Ram', 'Mohan', 'Tina', 'Jeetu', 'Meera'],  
        'Last_name': ['Kumar', 'Sharma', 'Ali', 'Gandhi', 'Kumari'],  
        'Age': [42, 52, 36, 21, 23],  
        'City': ['Mumbai', 'Noida', 'Pune', 'Delhi', 'Bihar']} 

info_df=pd.DataFrame(info, columns = ['First_name', 'Last_name', 
                                                      'Age', 'City']) 
info_df
```

```python
#Now lets add a new column 'qualifications' by mapping
qual_dict={ "Ram":"B.Com", 
             "Mohan":"IAS", 
             "Tina":"LLB", 
             "Jeetu":"B.Tech", 
             "Meera":"MBBS" } 
info_df['Qualification']=info_df['First_name'].map(qual_dict)
info_df
```

```python
#Lets use replace function to change some items in the City column
city_change={'Mumbai':'Washington','Noida':'Tokyo','Bihar':'Seoul'}
info_df=info_df.replace({'City':city_change})
info_df
```

```python
#Lets update the First_name using the index
name_update={0:'Shyam',3:'Joseph',4:'Messi'}
info_df['First_name'].update(pd.Series(name_update))
info_df
```
