```python
import numpy as np
import pandas as pd
#Adding New Columns in a Dataframe
#Lets create a DataFrame
countries=[('Nigeria',120,'Abuja','Africa'),
           ('Egypt',50,'Cairo','Africa'),
           ('Japan',110,'Tokyo','Asia'),
           ('India',1000,'Mumbai','Asia'),
           ('USA',700,'Washington','North America'),
           ('Brazil',500,'Rio','South America'),
           ('Australia',40,'Sydney','Oceania')
    
]
countries_df=pd.DataFrame(countries,columns=['Country','Population','Capital','Continent'],
                         index=['a','b','c','d','e','f','g'])
countries_df
```

```python
#Lets add a new column "Revenue" to it
countries_df['Revenue']=[124,89,578,782,900,850,80]
countries_df
#Note that if value provided in list are less than indexes,thats error
#If Column already exists,it will replace all the values there
```

```python
#Lets add a new column with a default value
countries_df['Crime']='Low'
countries_df
```

```python
#Lets add a calculated field-Percentage
countries_df['Percentage']=(countries_df['Population']/countries_df['Revenue'])*100
countries_df
```

```python
#Lets append column to dataframe using assign() keyword
#This accept the keyword:value pair, where column name is keyword and value could be list, series etc.
#Lets Reset Dataframe
countries_df2=pd.DataFrame(countries,columns=['Country','Population','Capital','Continent',])
countries_df2
```

```python
#Lets append GDP column to Dataframe
modified_df2=countries_df2.assign(GDP=[14.2,13.5,20.3,18.4,9.6,7.8,5.9])
modified_df2
```

```python
#Lets add multiple columns using the assign keyword
mod_df=countries_df2.assign(Crime=['High','Low','Low','High','Medium','Medium','Low'],
                           Color=['Black','White','Mongoloid','Yellow','White','Mixed','White'],
                           Fixed=[50]*7)
mod_df
```

```python
#Lets add a column based on other columns using lambda function
mod_df1=countries_df2.assign(Percentage=lambda x:(x['Population']/1000)*100)
mod_df1
```

```python
#HANDLING DUPLICATE COLUMNS
#Lets create a dataframe
info     = [('jack', 34, 'Sydeny'),
            ('Riti', 30, 'Delhi'),
            ('Aadi', 16, 'New York'),
            ('Riti', 30, 'Delhi'),
            ('Riti', 30, 'Delhi'),
            ('Riti', 30, 'Mumbai'),
            ('Aadi', 40, 'London'),
            ('Sachin', 30, 'Delhi')
            ]
info_df=pd.DataFrame(info,columns=['Name','Age','City'])
info_df
```

```python
dup_rows=info_df[info_df.duplicated()] #This returns duplicate rows except the 1st occurence
dup_rows
```

```python
dup_rows2=info_df[info_df.duplicated(keep='last')] #This returns duplicate rows except the last occurence
dup_rows2
```

```python
#We can also select duplicate columns based on one or more columns
dup_rows3=info_df[info_df.duplicated('Name')]
dup_rows3
```

```python
dup_rows4=info_df[info_df.duplicated(['Age','City'])]
dup_rows4
```

```python
#FIND MAXIMUM VALUES & POSITIONS IN COLUMNS OR ROWS
num    = [(22, 16, 23),
          (33, np.NaN, 11),
          (44, 34, 11),
          (55, 35, np.NaN),
          (66, 36, 13)
          ]
num_df=pd.DataFrame(num,index=list('abcde'),columns=list('xyz'))
num_df
```

```python
#Get maximum values of every column
max_val=num_df.max()
max_val
```

```python
#Get maximum values of every row
max_val2=num_df.max(axis=1)
max_val2
```

```python
#Get maximum values by column as well as nan values
max_val3=num_df.max(skipna=False) #Here, it picks the nan values as the maximum for each columns
max_val3
```

```python
#Get the maximum value for a single or selectec column
max_bycol=num_df['z'].max()
max_bycol
```

```python
#We can also write it like this
max_bycol1=num_df.max()['z']
max_bycol1
```

```python
max_bycol2=num_df[['x','y']].max()
max_bycol2
```

```python
#Get row index label or position of maximum values of every column
max_index=num_df.idxmax()
max_index
```

```python
#Get column names of maximum values of every row
max_index2=num_df.idxmax(axis=1)
max_index2
```
