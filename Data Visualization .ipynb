```python
#Data Visualization with Matplotlib and Pyplot

#Plotting a Bar Chart

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

youtube_views=[534,690,258,402,724,689,352]
days=[1,2,3,4,5,6,7]

plt.bar(days,youtube_views,label='Youtube Viewers')
plt.xlabel('days')
plt.ylabel('views')
plt.title('Bar Chart of Youtube Views')
plt.legend()
plt.savefig('Youtube Bar Chart.png',dpi=500)
plt.show()
```

```python
#Plotting two or more Bar Charts together

youyube_views=[534,690,258,402,724,689,352]
facebook_views=[123,342,700,305,406,648,325]
days=[1,2,3,4,5,6,7]

plt.bar(days,youtube_views,label='Youtube Viewers')
plt.bar(days,facebook_views,label='Facebook Viewers')
plt.xlabel('Days')
plt.ylabel('Views')
plt.title('Social Media Views Bar Chart')
plt.legend()
#plt.savefig('Social Media View Bar Charts.png',dpi=500)
plt.show()
```

```python
#Creating Multiple Plots

youtube_views=[534,690,258,402,724,689,352]
facebook_views=[123,342,700,305,406,648,325]
twitter_views=[202,209,176,415,824,389,550]
days=[1,2,3,4,5,6,7]

plt.plot(days,youtube_views,label='Youtube Viewers',marker='o') #This gives a line graph
plt.plot(days,facebook_views,label='Facebook Viewers',marker='o')
plt.plot(days,twitter_views,label='Twitter Viewers',marker='o')

plt.title('Social Media Views')
plt.xlabel('Days')
plt.ylabel('Views')
plt.legend()
plt.show()
```

```python
#Creating a Pie Chart

social_media=['Facebook','Instagram','Youtube','Linked-In']
views=[300,350,400,450]
exp=[0,0,0,0.2]

plt.pie(views,labels=social_media)
plt.savefig('Social Media Pie Chart.png')
plt.show()

#We can add some other parameters like
#autopct......used to get percentages
#explode
#shadow
```

```python
#Data Visualization Using seaborn
from numpy.random import randn
import seaborn as sns

#Creating Box Plots...useful for doing stock market and financial analysis
dataset1=randn(80)
#dataset1
sns.boxplot(dataset1,color='blue').get_figure()#.savefig('BoxPlot1.png')
```

```python
#USING THE ORIENT FUNCTION TO CHANGE THE ORIENTATION
sns.boxplot(dataset1,color='green',orient='v').get_figure()#.savefig('Boxplot2.png')
```

```python
#USING WHIS FUNCTON TO TAKE CARE OF OUTLAYER POINT
sns.boxplot(dataset1,color='purple',whis=np.inf).get_figure()#.savefig('BoxPlot3')
```

```python
#Creating Cluster Maps
#Note,the clustermap plot doesnt have the get_figure function
passenger_data=sns.load_dataset('flights')
print (passenger_data)
```

```python
passenger_df=passenger_data.pivot('year','month','passengers')
passenger_df
```

```python
sns.clustermap(passenger_df)#.savefig('Clustermap1')
```

```python
#Clustering in one direction
sns.clustermap(passenger_df,col_cluster=False)#.savefig('Clustermap2')
```

```python
#Standadizing by rows
sns.clustermap(passenger_df,standard_scale=0)#.savefig('Clustermap3')
```

```python
#Standadizing by columns
sns.clustermap(passenger_df,standard_scale=1)#.savefig('Clustermap4')
```

```python
#Creating Heatmaps
sns.heatmap(passenger_df).get_figure()#.savefig('HeatmapPlot1')
```

```python
#Using Annotations
sns.heatmap(passenger_df,annot=True,fmt='d').get_figure()#.savefig('HeatmapPlot2')
```

```python
#Use of Center Location...This can be used to easily predict profit or loss
sns.heatmap(passenger_df,center=passenger_df.loc[1955,'January']).get_figure()#.savefig('HeatmapPlot3')
```
