```python
#DATAFRAME MODIFICATION
import numpy as np
import pandas as pd
matrix = [(222, 34, 23),
         (333, 31, 11),
         (444, 16, 21),
         (555, 32, 22),
         (666, 33, 27),
         (777, 35, 11)
         ]
mat_df=pd.DataFrame(matrix,columns=list('ABC'))
mat_df
```

```python
#Lets apply a lambda function to each row or column
#Columns
mod_df=mat_df.apply(lambda x:x+10)
mod_df
```

```python
#Rows
mod_df1=mat_df.apply(lambda x:x+5, axis=1)
mod_df1
```

```python
#Applying User defined functions with or without arguments to dataframe
def doubledata(x):
    return x*2
mod_df2=mat_df.apply(doubledata)
mod_df2
```

```python
#we can also apply it to each row 
mod_df3=mat_df.apply(doubledata,axis=1)
mod_df3
```

```python
#Applying a user defined function with arguments
def multiplydata(x,y):
    return x*y
mod_df4=mat_df.apply(multiplydata,args=[3])
mod_df4
```

```python
#Similarly, we can apply it to rows
mod_df5=mat_df.apply(multiplydata,axis=1,args=[4])
mod_df5
```

```python
#Applying numpy function to each row or column
#1.square function
sqr_df=mat_df.apply(np.square)
sqr_df
```

```python
sqr_df1=mat_df.apply(np.square, axis=1)
sqr_df1
```

```python
#2.sum function
sum_df=mat_df.apply(np.sum) #This sums up each columns
sum_df
```

```python
sum_df1=mat_df.apply(np.sum, axis=1)
sum_df1
```

```python
#Applying a function to a single or selected columns or rows
rand_num=[(12,10,14),
          (8,4,2),
          (11,22,10),
          (5,4,21),
          (20,15,5)
         ]
rand_df=pd.DataFrame(rand_num,columns=list('XYZ'),index=list('abcde'))
rand_df
```

```python
#Apply function to a single column
#Method 1-DataFrame.apply()-let us square column Z
rand1=rand_df.apply(lambda x:np.square(x) if x.name=='Z' else x)
rand1
```

```python
#Method 2-Using [] operator
rand_df['Z']=rand_df['Z'].apply(np.square)
rand_df
```

```python
#Method 3-Using np.square
rand_df['Z']=np.square(rand_df['Z'])
rand_df
```

```python
#Lets apply a function to a selected row
#Method 1-DataFrame.apply()
rand2=rand_df.apply(lambda x:x*2 if x.name=='b' else x, axis=1)
rand2
```

```python
#Method 2-Using the [] operator
rand2.loc['b']=rand2.loc['b'].apply(np.square)
rand2
```

```python
#Method 3-Using numpy.square
rand2.loc['d']=rand2.loc['d'].apply(np.square)
rand2
```

```python
#In the same vein, we can apply functions to multiple rowS and columns
matric_num=[(3,3,6,8),
            (5,4,7,10),
            (7,8,9,5)
    
]
matric_df=pd.DataFrame(matric_num,columns=list('abcd'))
matric_df
```

```python
#Multiple columns
matric1=matric_df.apply(lambda x:np.square(x) if x.name in ['a','c'] else x)
matric1
```

```python
#Multiple rows
matric2=matric_df.apply(lambda x:np.square(x) if x.name in [0,1] else x, axis=1)
matric2
```

```python
#Sorting Rows and Columns in Dataframe based on values
students = [ ('Jack', 34, 'Sydney') ,
             ('Riti', 41, 'Delhi' ) ,
             ('Aadi', 16, 'New York') ,
             ('Riti', 22, 'Delhi' ) ,
             ('Riti', 35, 'Delhi' ) ,
             ('Riti', 40, 'Mumbai' )
              ]
students_df=pd.DataFrame(students,columns=['Name','Age','City'],index=['b','a','f','e','d','c'])
students_df
```

```python
#Sort dataframe rows based on a single column
sort1=students_df.sort_values(by='Name')
sort1
```

```python
#Sort dataframe based on multiple columns
sort2=students_df.sort_values(by=['Name','Age'])
sort2
```

```python
#Lets sort based on a column in descending order
sort3=students_df.sort_values(by='Name',ascending=False)
sort3
```

```python
#Sort dataframe rows based on a column in place
students_df.sort_values(by='Name',inplace=True) #This takes effect on dataframe
students_df
```

```python
#Sort dataframe columns based on a single or multiple rows
matr    = [(222, 16, 23),
          (333, 31, 11),
          (444, 34, 11),
          ]
matr_df=pd.DataFrame(matr,index=list('abc'))
matr_df
```

```python
#Sort columns based on a single row
mod1=matr_df.sort_values(by='b',axis=1)
mod1
```

```python
#sort columns based on a single row in descending order
mod2=matr_df.sort_values(by='b',axis=1,ascending=False)
mod2
```

```python
#Sort columns based on multiple rows
mod3=matr_df.sort_values(by=['b','c'],axis=1)
mod3
```

```python
#Sorting using the DataFrame.sort_index for rows and columns
cars=[('venza',5,'japan'),
      ('benz',8,'germany'),
      ('audi',4,'uae'),
      ('royce',8,'usa'),
      ('gwagon',9,'italy'),
      ('innoson',3,'nigeria')
    
]
cars_df=pd.DataFrame(cars,columns=['Name','Price','Country'],index=['b','a','f','e','c','d'])
cars_df
```

```python
#Sort rows of dataframe based on row index labels
cars_df1=cars_df.sort_index()
cars_df1
```

```python
#Sort rows of dataframe based on row index in descending
cars_df2=cars_df.sort_index(ascending=False)
cars_df2
#Note that we can modify the dataframe by passing the inplace=True argument
```

```python
#Sort columns of Dataframe based on column names
cars_df3=cars_df.sort_index(axis=1)
cars_df3
#We can sort in ascending order by passing ascending=False argument
#We can also modify the dataframe by passing inplace=True argument
```

```python
#Convert Dataframe column into index
emp_sal  = [('jack', 34, 'Sydney', 70000),
            ('Riti', 31, 'Delhi', 77000),
            ('Aadi', 16, 'Mumbai', 81000),
            ('Mohit', 31, 'Delhi', 90000),
            ('Veena', 12, 'Delhi', 91000),
            ('Shaunak', 35, 'Mumbai', 75000),
            ('Mark', 35, 'Colombo', 63000)
            ]
sal_df=pd.DataFrame(emp_sal,columns=['Name','Age','City','Salary'])
sal_df
```

```python
#Give the row index a name
sal_df.index.rename('Id',inplace=True)
sal_df
```

```python
#Lets make column 'Name' the index
mod_sal=sal_df.set_index('Name')
mod_sal
```

```python
#We can also make column 'Name' the index without removing the column
mod_sal2=sal_df.set_index('Name',drop=False)
mod_sal2
```

```python
#Let us Append column 'Name' to index to make it multi-index
mod_sal3=sal_df.set_index('Name',append=True)
mod_sal3
```

```python
#Check for duplicates in new index
mod_sal4=sal_df.set_index('City',verify_integrity=True) #This will return an error,index has duplicate keys.
mod_sal4
```

```python
#Convert Dataframe Index to columns
emp_id = [(11, 'jack', 34, 'Sydney', 70000) ,
          (12, 'Riti', 31, 'Delhi' , 77000) ,
          (13, 'Aadi', 16, 'Mumbai', 81000) ,
          (14, 'Mohit', 31,'Delhi' , 90000) ,
          (15, 'Veena', 12, 'Delhi' , 91000) ,
          (16, 'Shaunak', 35, 'Mumbai', 75000 ),
          (17, 'Shaun', 35, 'Colombo', 63000)
           ]
id_df=pd.DataFrame(emp_id,columns=['ID','Name','Age','City','salary'])
id_df
```

```python
#Convert 'ID' to index
id_df.set_index('ID',inplace=True)
id_df
```

```python
#Convert index 'ID' back to a Column
#id_df=id_df.reset_index() or
id_df.reset_index(inplace=True)
id_df
```

```python
#Making 'ID' the index
id_df.set_index('ID',inplace=True)
id_df
```

```python
#Remove Index 'ID'
mod_id=id_df.reset_index(drop=True)
mod_id
```

```python
#Creating a multi-index dataframe
id_df2=pd.DataFrame(emp_id,columns=['ID','Names','Age','City','Salary'])
id_df2
```

```python
id_df2.set_index(['ID','Names'],inplace=True)
```

```python
id_df2
```
